```python
import numpy as np
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('country_full.csv')
data
```

```python
# data.head(50)
```

```python
# data.tail(50)
```

```python
data.drop(['alpha-2','alpha-3','iso_3166-2','intermediate-region','intermediate-region-code'],axis=1,inplace=True)
data
```

```python
reddy=data[data.region =='Europe']
```

```python
plt.barh(reddy['name'],reddy['country-code'],color='r')
plt.title('Asia	sub-region-code')
plt.xlabel('region code')
plt.gcf().set_size_inches(50,40)
plt.show()
```

```python
data
```

```python
display(data['region'].value_counts())
```

```python
Country=data['region'].value_counts().head()
name=data['region'].value_counts().head().index
plt.gcf().set_size_inches(50,38)
plt.pie(Country,labels=name,autopct='%1.f%%',wedgeprops={"edgecolor" :"black"},explode=[0,.1,.2,.1,0])
plt.title("top country code")
plt.show()
```

```python
Country=data['region'].value_counts().head(2)
name=data['region'].value_counts().head(2).index
plt.gcf().set_size_inches(50,38)
plt.pie(Country,labels=name,autopct='%1.f%%',wedgeprops={"edgecolor" :"r"},explode=[.1,.2])
plt.title("top country code")
plt.show()
```

```python
reddy=data[data.region =='Oceania']
```
